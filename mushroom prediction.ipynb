from the above diagram we cannot say which one is more related with label so we use heat map and remove one feature between two

```python
#veil_color is more related comparing with gill attachment so we remove gill_attachment
x=new_df.drop(columns=['gill_attachment','class'])
y=new_df['class']
```

```python
x.shape
```

```python
#standardizing the data
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
x_scaled=scalar.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.3,random_state=30)
```

# logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg=LogisticRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
y_pred=lg.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,roc_curve,roc_auc_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

# KNN Model

```python
#applying knn model 
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
#accuracy score is 1.0 it is looks like our model is overfitting so we check if it is over
#fitting or not by using cross validation score
from sklearn.model_selection import cross_val_score
cross_val_score(knn,x_scaled,y,cv=10)
```

```python
cross_val_score(KNeighborsClassifier(),x_scaled,y,cv=5).mean()
```

we absorve that from cross validation accuracy score is drastically different so we confirm that our model is overfitting 

we use hyperparameter tunning for improve accuracy

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'algorithm':['kd_tree','brute'],'leaf_size':[1,5,8,4,2],'n_neighbors':range(1,5)}
```

```python
gridsearch=GridSearchCV(estimator=knn,param_grid=param_grid)
```

```python
gridsearch.fit(x_train,y_train)
```
