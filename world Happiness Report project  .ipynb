```python
x=new_df.drop(columns=['Score'])
y=new_df['Score']
```

```python
#standardization of data
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
x_scaled=scalar.fit_transform(x)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=10)
```

```python
#training the data
regression=LinearRegression()
regression.fit(x_train,y_train) 
```

```python
regression.score(x_train,y_train)#accuracy score on training data
```

```python
regression.score(x_test,y_test)#accuracy score on test data
```

```python
y_pred=regression.predict(x_test)
y_pred
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel('actual Happyness Score')
plt.ylabel('predicted Happyness Score')
plt.title('actual vs model prediction')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
y_pred=regression.predict(x_test)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
#check if the model is over ffitting or not by using regularization techniques LASSO and RIDGE
from sklearn.linear_model import Ridge,Lasso,RidgeCV,LassoCV
```

```python
lascv=LassoCV(alphas=None,max_iter=50,normalize=True)#lassoCv is giving best alpha(learning rate) after max_iterations
lascv.fit(x_train,y_train)
```

```python
alpha=lascv.alpha_
alpha
```

```python
las_lnrgsn=Lasso(alpha)
las_lnrgsn.fit(x_train,y_train)
```

```python
las_lnrgsn.score(x_test,y_test)
```
